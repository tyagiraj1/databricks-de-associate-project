```python
# Cell 1 - Define all paths
source_path      = "/Volumes/workspace/default/raw_orders/"
bronze_table     = "workspace.default.bronze_orders"
checkpoint_path  = "/Volumes/workspace/default/raw_orders/_checkpoints/bronze"
schema_path      = "/Volumes/workspace/default/raw_orders/_schema"
```

```python
# Cell 2 - Read stream using Autoloader (cloudFiles)
df_raw = (
    spark.readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "json")
    .option("cloudFiles.schemaLocation", schema_path)
    .option("cloudFiles.inferColumnTypes", "true")
    .load(source_path)
)
```

```python
# Cell 3 - Write stream to Bronze Delta table
(
    df_raw.writeStream
    .format("delta")
    .option("checkpointLocation", checkpoint_path)
    .outputMode("append")
    .trigger(availableNow=True)
    .toTable(bronze_table)
)
```
