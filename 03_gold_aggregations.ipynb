```python
%sql
-- Cell 1 - Total sales and orders by category
CREATE OR REPLACE TABLE workspace.default.gold_sales_by_category AS
SELECT
    category,
    COUNT(order_id)        AS total_orders,
    SUM(quantity)          AS total_units_sold,
    ROUND(SUM(total_amount), 2) AS total_revenue,
    ROUND(AVG(total_amount), 2) AS avg_order_value
FROM workspace.default.silver_orders
GROUP BY category
ORDER BY total_revenue DESC
```

```python
%sql
-- Cell 2 - Total sales and orders by city
CREATE OR REPLACE TABLE workspace.default.gold_sales_by_city AS
SELECT
    city,
    COUNT(order_id)             AS total_orders,
    SUM(quantity)               AS total_units_sold,
    ROUND(SUM(total_amount), 2) AS total_revenue,
    ROUND(AVG(total_amount), 2) AS avg_order_value
FROM workspace.default.silver_orders
GROUP BY city
ORDER BY total_revenue DESC
```

```python
%sql
-- Cell 3 - Monthly sales trend
CREATE OR REPLACE TABLE workspace.default.gold_monthly_trend AS
SELECT
    DATE_FORMAT(order_date, 'yyyy-MM')  AS order_month,
    COUNT(order_id)                     AS total_orders,
    SUM(quantity)                       AS total_units_sold,
    ROUND(SUM(total_amount), 2)         AS total_revenue
FROM workspace.default.silver_orders
GROUP BY DATE_FORMAT(order_date, 'yyyy-MM')
ORDER BY order_month ASC
```

```python
%sql
-- Cell 4 - Top 5 products by revenue
CREATE OR REPLACE TABLE workspace.default.gold_top_products AS
SELECT
    product,
    category,
    COUNT(order_id)             AS total_orders,
    SUM(quantity)               AS total_units_sold,
    ROUND(SUM(total_amount), 2) AS total_revenue
FROM workspace.default.silver_orders
GROUP BY product, category
ORDER BY total_revenue DESC
LIMIT 5
```
